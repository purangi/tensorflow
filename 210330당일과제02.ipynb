```python
import tensorflow as tf
import random

x1 = [ 73., 93., 89., 96., 73. ]
x2 = [ 80., 88., 91., 98., 56. ]
x3 = [ 75., 93., 90., 100.,70. ]
Y = [152., 185., 180., 196., 142. ]

#초기값을 1로 설정(주로 랜덤값)
w1 = tf.Variable(tf.random.normal([1]))
w2 = tf.Variable(tf.random.normal([1]))
w3 = tf.Variable(tf.random.normal([1]))
b = tf.Variable(tf.random.normal([1]))
learning_rate = 0.000001

#10001번의 순회
for i in range(10000+1):
  with tf.GradientTape() as tape:
    hypothesis = w1 + x1 + w2 + x2 + w3 + x3 + b
    cost = tf.reduce_mean(tf.square(hypothesis - Y)) #오차의 제곱의 평균 구하기

  w1_grad, w2_grad, w3_grad, b_grad = tape.gradient(cost, [w1, w2, w3, b]) #기울기 구하기
  w1.assign_sub(learning_rate * w1_grad) #개별 업데이트
  w2.assign_sub(learning_rate * w2_grad)
  w3.assign_sub(learning_rate * w3_grad)
  b.assign_sub(learning_rate * b_grad)
  if i % 50 == 0:
    print("{:5} | {:12.4f}".format(i, cost.numpy())) #cost값 출력
```
